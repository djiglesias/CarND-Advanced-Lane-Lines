## 1.0 Pipeline (Single Image)

```python
import cv2
import glob
import matplotlib.pyplot as plt

from main.lane import Lane

%matplotlib inline

# Load test videos from file.
files = glob.glob('./test_images/*.jpg')

# Apply pipeline to video.
lane = Lane()
filename = files[0]
image = lane.process(cv2.imread(filename))

# Plot axes.
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
